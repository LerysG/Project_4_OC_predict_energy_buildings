## Variable définitions :


- Site = "https://www.splitgraph.com/seattle-gov/2016-building-energy-benchmarking-2bpz-gwpy/latest/-/api-schema"

- Abréviations :
- **WN** : Weather Normalized
- **EUI** : Energy Use Intensity
- **sf** : Squared Feet

- Définitions :
- **SourceEUI** : Source Energy Use Intensity (EUI) is a property's Source Energy Use divided by its gross floor area. Source Energy Use is the annual energy used to operate the property, including losses from generation, transmission, & distribution.

- **SiteEUI** : Site Energy Use Intensity (EUI) is a property's Site Energy Use divided by its gross floor area. Site Energy Use is the annual amount of all the energy consumed by the property on-site, as reported on utility bills

- **LargestPropertyUseTypeGFA** : The gross floor area (GFA) of the largest use of the propert

- **DefaultData** : The property used default data for at least one property characteristic

- **PropertyGFAtotal** : Total building and parking gross floor area.

```python
df = pd.read_csv(PATH_df)
df = df.dropna()
df.drop(columns=['Unnamed: 0', 'index'], inplace=True)
df['CouncilDistrictCode'] = df['CouncilDistrictCode'].apply(str)
df2015 = df[df['DataYear'] == 2015]
df2016 = df[df['DataYear'] == 2016]
```

# Train-Test split

```python
X = df2015[["BuildingType",
        "PrimaryPropertyType",
        "CouncilDistrictCode",
        "Neighborhood",
        "YearBuilt",
        "NumberofBuildings",
        "NumberofFloors",
        "LargestPropertyUseType",
        "PropertyGFAParking",
        "LargestPropertyUseTypeGFA",
        "SteamUse(kBtu)_percentage",
        "Electricity(kBtu)_percentage",
        "NaturalGas(kBtu)_percentage",
        "ENERGYSTARScore"
        ]]
y = np.log(df2015['SiteEnergyUse_kBtu'])                         
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=13)
```

# Preprocessor

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, RobustScaler, OneHotEncoder, OrdinalEncoder

numeric_features = X_train.select_dtypes(include = np.number).columns
RS = RobustScaler()

categorical_features = X_train.select_dtypes(exclude = np.number).columns
OE = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
OHE = OneHotEncoder(handle_unknown='ignore')
                                           
preprocessor_OE = ColumnTransformer(
    transformers=[
        ('num', RS, numeric_features),
        ('cat', OE, categorical_features)])
                                          
preprocessor_OHE = ColumnTransformer(
    transformers=[
        ('num', RS, numeric_features),
        ('cat', OHE, categorical_features)])
```

# Dimensionality reduction (truncated SVD)

```python
from sklearn.decomposition import PCA, TruncatedSVD
X_train_scaled = preprocessor_OHE.fit_transform(X_train)
model = TruncatedSVD(n_components=40)
X_train_scaled_reduced = model.fit_transform(X_train_scaled)
variance = np.cumsum(model.explained_variance_ratio_)

plt.plot(np.linspace(1,40,40), variance)
plt.xlabel('Number of Principal Component')
plt.ylabel('Variance (%)')


np.argmax(variance > 0.98)
```

```python
y_train.shape
X_train_scaled.shape
model.fit_transform(X_train_scaled).shape
```

```python
model = TruncatedSVD(n_components=2)
principalComponents = model.fit_transform(X_train_scaled)
principalDf = pd.DataFrame(data = principalComponents , columns = ['component_1', 'component_2'])
plt.figure(figsize=(10,10))
sns.scatterplot(data= principalDf, x = 'component_1', y = 'component_2', 
                size = y_train, sizes=(1, 200) 
               )
```

# Linear Regression

## Baseline

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

regressor = LinearRegression()

model = Pipeline(steps=[('preprocessor', preprocessor_OHE),
                        ('regressor', regressor)
                        ])

param_grid = {}

grid = GridSearchCV(model,
                    param_grid,
                    cv=5,
                    scoring='r2')

grid.fit(X_train, (y_train))

print(round(grid.best_score_, 3))
print(grid.best_params_)

model.fit(X_train, y_train)
model.score(X_train, y_train)
model.score(X_test, y_test)

model.fit(X_train, 10**(y_train))
model.score(X_train, 10**(y_train))
model.score(X_test, 10**(y_test))
```

## Linear Regression Lasso 

```python
from sklearn.linear_model import Lasso
n_alphas = 60
alphas = np.logspace(-5, 5, 11)

coefs = []
errors = []

lasso = Lasso()

processed_X = preprocessor_OHE.fit_transform(X)
X_train2, X_test2, y_train2, y_test2 = train_test_split(processed_X, y, test_size = 0.2, random_state=13)

for a in alphas:
    lasso.set_params(alpha=a)
    lasso.fit(X_train2, y_train2)
    coefs.append(lasso.coef_)
    errors.append(np.mean((lasso.predict(X_test2) - y_test2) **2))
```

```python
plt.figure()
plt.plot(alphas, errors)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('alpha')
plt.ylabel('errors')
plt.legend([ "errors"])
plt.show()
    
plt.figure()
plt.plot(alphas, coefs)
plt.xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefs')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

regressor = Lasso(max_iter=5000)

model = Pipeline(steps=[('preprocessor', preprocessor_OHE),
                        ('regressor', regressor)
                        ])

param_grid = {'regressor__alpha': np.logspace(-5, 5, 11)}

grid = GridSearchCV(model,
                    param_grid,
                    cv=5,
                    scoring='r2')

grid.fit(X_train, (y_train))

print(round(grid.best_score_, 3))
print(grid.best_params_)
```

```python
model = grid.best_estimator_

model.fit(X_train, y_train)
model.score(X_train, y_train)
model.score(X_test, y_test)
```

## Testset score

```python
preprocessor_OHE.named_transformers_['num'].fit(X_train.select_dtypes(include=np.number))
preprocessor_OHE.named_transformers_['cat'].fit(X.select_dtypes(exclude=np.number))

X_train_scaled = preprocessor_OHE.transform(X_train)
X_test_scaled = preprocessor_OHE.transform(X_test)

model =  grid.best_estimator_.named_steps['regressor']
model.fit(X_train_scaled, y_train)
score = model.score(X_test_scaled, y_test)
y_pred = model.predict(X_test_scaled)

plt.figure(figsize=(7,7))

plt.subplot(211)
sns.regplot(x = y_test, y = y_pred, label='test vs. predicted', color='black')
plt.plot(y_test,y_test, '--', label='x=y', color ='r', linewidth = 3)
plt.xlabel('Energy use (test set, log)')
plt.ylabel('Energy use (predicted, log)')
plt.legend()


plt.subplot(212)
sns.regplot(x = y_test, y = (100*(y_pred-y_test)/y_test), label='Residual', color='black')
plt.xlabel('Energy use (test set, log)')
plt.ylabel('Residual (%)')
plt.ylim(-100,100)
plt.legend()

plt.tight_layout(pad=0.5)
plt.show()
print(f"R2 = {score:.3f}")
```

## Feature importance

```python
from sklearn.inspection import permutation_importance

# Recover features columns (especially after OneHotEncoder)
onehot_columns = list(preprocessor_OHE.named_transformers_[
                      'cat'].get_feature_names(input_features=categorical_features))
numeric_features_list = list(numeric_features)
features_list = np.concatenate((numeric_features_list, onehot_columns))

# Train model with best estimator from previous GridSearchCV
model = grid.best_estimator_.named_steps['regressor']
model.fit(X_train_scaled, y_train)

# Recover important features from previous fit and then permute importance
n_feat = 10  # number of feature to show

feature_importance = model.feature_importances_

sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5

fig = plt.figure(figsize=(20, 6))

plt.subplot(1, 2, 1)
plt.barh(pos[-1*n_feat:], feature_importance[sorted_idx]
         [-1*n_feat:], align='center')
plt.yticks(pos[-1*n_feat:], features_list[sorted_idx][-1*n_feat:])
plt.title('Feature Importance (MDI)')

result = permutation_importance(model, X_test_scaled.toarray(), y_test, n_repeats=10,
                                random_state=42, n_jobs=2)
sorted_idx = result.importances_mean.argsort()[-1*n_feat:]

plt.subplot(1, 2, 2)
plt.boxplot(result.importances[sorted_idx].T[-1*n_feat:],
            vert=False, labels=features_list[sorted_idx][-1*n_feat:])
plt.title("Permutation Importance (test set)")
fig.tight_layout(pad=0.8)
plt.show()
```

## Validation Curve

```python
from sklearn.model_selection import validation_curve
model = grid.best_estimator_.named_steps['regressor']
n = np.linspace(100, 1000, 5).astype(int)

train_score, val_score = validation_curve(
    model, X_train_scaled, (y_train), 'n_estimators', n, cv=5)


plt.plot(n, train_score.mean(axis=1), label='train set', color ="dodgerblue", linewidth = 2)
plt.plot(n, val_score.mean(axis=1), label='validation set', color ="gold", linewidth = 2)

plt.xlabel('n')
plt.ylabel('R2')
plt.ylim(0.5,1)
plt.legend()
plt.title("Validation Curve")

```

## Learning Curve
